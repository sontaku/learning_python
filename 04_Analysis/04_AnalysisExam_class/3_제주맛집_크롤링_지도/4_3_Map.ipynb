```python
# ì˜ˆì œ 4-26  ì¥ì†Œ ê²€ìƒ‰ í•¨ìˆ˜ ë§Œë“¤ê¸°
def find_places(searching):
    # â‘  ì ‘ì†URL ë§Œë“¤ê¸°
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    # â‘¡ headers ì…ë ¥í•˜ê¸°
    headers = {
    # "Authorization": "KakaoAK ì—¬ê¸°ì— ë³¸ì¸ì˜ ì¹´ì¹´ì˜¤APIí‚¤ì…ë ¥í•˜ì„¸ìš”"     
    "Authorization": "KakaoAK c365fabe892788f00e612734bd99c75f" # 
    # ì…ë ¥ì‹œ ë°˜ë“œì‹œ  KakaoAK ë’¤ì— í•œ ì¹¸ ë„ìš°ê³  APIí‚¤ ê°’ì„ ì ì–´ì•¼ í•©ë‹ˆë‹¤. 
    }
    # â‘¢ API ìš”ì²­&ì •ë³´ ë°›ê¸°
    places = requests.get(url, headers = headers).json()['documents']
    # â‘£ í•„ìš”í•œ ì •ë³´ ì„ íƒí•˜ê¸°
    place = places[0] 
    name = place['place_name']
    x=place['x']
    y=place['y']
    data = [name, x, y, searching] 

    return data
```

```python
# ì˜ˆì œ 4-27 ì œì£¼ê³µí•­ ê²€ìƒ‰ ì˜ˆì‹œ
data = find_places('ì œì£¼ê³µí•­')
data
```

```python
from tqdm import tqdm_notebook
```

```python
# ì˜ˆì œ 4-28 ì¸ìŠ¤íƒ€ê·¸ë¨ ìœ„ì¹˜ëª… ìœ„ì¹˜ì •ë³´ ê²€ìƒ‰í•˜ê¸°
#### ê²€ìƒ‰ì‹œ ì‹œê°„ì´ ë§ì´ ê±¸ë ¤, í˜„ì¬ ì‘ì—… ì§„í–‰í˜„í™©ì„ ë³¼ ìˆ˜ ìˆë„ë¡,  tqdm_notebook ì¶”ê°€ í–ˆìŠµë‹ˆë‹¤.
import time
locations_inform = [ ]
for location in tqdm_notebook(locations):
    try:
        data = find_places(location)       
        locations_inform.append(data) 
        time.sleep(0.5) 
    except:
        pass
locations_inform
```

```python
# ì˜ˆì œ 4-30 ì¸ìŠ¤íƒ€ ê²Œì‹œëŸ‰ ë° ìœ„ì¹˜ì •ë³´ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
location_counts_df = pd.read_excel('./files/3_3_location_counts.xlsx', index_col = 0)
locations_inform_df = pd.read_excel('./files/3_3_locations.xlsx')
```

```python
# ì˜ˆì œ 4-31 ìœ„ì¹˜ ë°ì´í„° ë³‘í•©í•˜ê¸°
location_data = pd.merge(locations_inform_df, location_counts_df, 
                         how = 'inner', left_on = 'name_official', right_index=True)

location_data.head()
```

### 4.3.6 foliumì„ ì´ìš©í•œ ì§€ë„ ì‹œê°í™” â‘  - ê°œë³„ í‘œì‹œ

```python
# ì˜ˆì œ 4-33 ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
location_data = pd.read_excel('./files/3_3_location_inform.xlsx')
location_data.info()
```

```python
# ì˜ˆì œ 4-34 ì§€ë„ í‘œì‹œí•˜ê¸°

import folium

Mt_Hanla =[33.362500, 126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11)

for i in range(len(location_data)):
    name = location_data ['name_official'][i]    # ê³µì‹ëª…ì¹­
    count = location_data ['place'][i]           # ê²Œì‹œê¸€ ê°œìˆ˜
    size = int(count)*2
    long = float(location_data['ìœ„ë„'][i])      
    lat = float(location_data['ê²½ë„'][i])       
    folium.CircleMarker((long,lat), radius = size, color='red', popup=name).add_to(map_jeju)
    
map_jeju
```

```python
# ì˜ˆì œ 4-35 ì§€ë„ ì €ì¥í•˜ê¸°
map_jeju.save('./files/3_3_jeju.html') 
```

### 4.3.7 foliumì„ ì´ìš©í•œ ì§€ë„ ì‹œê°í™” â‘¡ - ê·¸ë£¹ìœ¼ë¡œ í‘œì‹œ

```python
# ì˜ˆì œ 4-36 ì§€ë„ í‘œì‹œí•˜ê¸°(ë§ˆì»¤ ì§‘í•©)

from folium.plugins import MarkerCluster

locations = []
names = []

for i in range(len(location_data)):
    data = location_data.iloc[i]  # í–‰ í•˜ë‚˜ì”©
    locations.append((float(data['ìœ„ë„']),float(data['ê²½ë„'])))    # ìœ„ë„ , ê²½ë„ ìˆœìœ¼ë¡œ..
    names.append(data['name_official'])

icon_create_function = """\
function(cluster) {
    return L.divIcon({
    html: '<b>' + cluster.getChildCount() + '</b>',
    className: 'marker-cluster marker-cluster-large',
    iconSize: new L.Point(30, 30)
    });
}"""

Mt_Hanla =[33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 11)
                       
marker_cluster = MarkerCluster(
    locations=locations, popups=names,
    name='Jeju',
    overlay=True,
    control=True,
    icon_create_function=icon_create_function
)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
```

```python
# ì˜ˆì œ 4-37 ì§€ë„ ì €ì¥í•˜ê¸°
map_jeju2.save('./files/3_3_jeju_cluster.html') 
```
