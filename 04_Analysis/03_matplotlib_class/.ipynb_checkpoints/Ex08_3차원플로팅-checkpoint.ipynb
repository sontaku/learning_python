```python
""" 1. 3차원 플로팅 """
%matplotlib inline
# %matplotlib notebook

""" %matplotlib inline 이 아닌 %matplotlib notebook 하면
    3차원 플로팅을 대화식 그림으로 볼 수 있다.
    대화식 그림이란 바로 바로 변경 가능한 그림
"""

import numpy as np
import matplotlib.pyplot as plt

# fig = plt.figure()
# ax = plt.axes()

# mplot3d 툴킷을 임포트하고 projection='3d'로 3차원 축을 만든다
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
""" 2. 3차원 점과 선"""
ax = plt.axes(projection='3d')

zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# 3차원 산점도를 위한 데이타
zdata = 15*np.random.random(100)
xdata = np.sin(zdata)+0.1 * np.random.randn(100)
ydata = np.cos(zdata)+0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')
```
