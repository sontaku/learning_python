```python
# 기본적인 그래프
import matplotlib.pyplot as plt

data = [10, 15, 20, 30, 25]

plt.plot(data)
```

```python
# 1- 그래프 틀(영역)을 만들기
# 2- 그래프 그리기
# 3- 그래프 보여주기

plt.figure( figsize=(12, 8) ) # 단위는 inch
plt.plot(data)
plt.show()
```

```python
# x축과 y축의 라벨 추가
plt.plot(data)
plt.xlabel('time')
plt.ylabel('grade')
plt.title('My chart')
```

1- 그래프 틀(영역)을 만들기
   > plt.figure()       
    
2- 그래프 그리기
   > plt.plot(data)     
    
3- 그래프 보여주기
   > plt.show()         

 
 ### *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다*

```python
# y축 데이타에 x축 데이타를 지정
plt.plot(data)
plt.grid() # 격자무늬 추가
```

```python
# 속성 : linestyle , color , marker
plt.plot(data, linestyle='dashed', color='red', marker='o');
```

```python
# 속성 축약
plt.plot(data, 'ro--');
```
