```python
%matplotlib inline
    
import matplotlib.pyplot as plt
data = [10,15, 19,30, 20]
plt.plot(data)

# 여기서 data는 y축이다. x축은 자동으로 지정됨
```

```python
# 기본적인 코딩하면 - 주피터아닌 곳에서 실행하려면 ( 최근엔 주피터 아닌 곳에서 테스트 안해 봄)

plt.figure()       # 1- 그래프 틀(영역)을 만들기
plt.plot(data)     # 2- 그래프 그리기
plt.show()         # 3- 그래프 보여주기

plt.figure(figsize=(12,8))       # 단위가 inch라는 것인가
plt.plot(data)     
plt.show() 
```

```python
# x축과 y축의 라벨 추가
plt.figure();
plt.plot(data);
plt.title('Example')   # 우선 영어로 작성한다 -> 한글은 안된다
plt.xlabel('time')
plt.ylabel('Amplitudue')
plt.show()
```

## *그래프 출력시 객체표시 같은 거 안 출력하게 하려면 뒤에 세미콜론(;) 붙이면 된다

```python
# y축 데이타에 x축 데이타를 지정

step = ['one','two','three','four','five']
plt.plot(step, data);   # <---------- 세미콜론
plt.grid()      # show() 전이라 출력되나?

'''
가급적 전문가들 코드를 보면 

plt.figure()       # 1- 그래프 틀(영역)을 만들기
plt.plot(data)     # 2- 그래프 그리기
plt.show()         # 3- 그래프 보여주기
'''
```

```python
# 속성 : linestyle , color , marker
plt.plot(step, data, color='blue', linestyle='dashed', marker='o');
```

```python
plt.plot(step, data,'ro--');
# 이렇게 축약해서 사용할 수 있다는 것만 확인
```

[참고] 선의 두께와 선의 색을 지정

- lw의 기본값은 1


- 선 색
    {'b','g','r','c','m','y', 'k', 'w' }

```python
plt.plot(step, data, lw=3, color='green', linestyle='dashed', marker='o');
```

```python
# 마커속성 
plt.plot(step, data, lw=3, color='green', linestyle='dashed', marker='o', markerfacecolor='cyan', markersize=12);
```
