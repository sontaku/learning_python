## 데이타 필터링

분석에 필요한 데이타를 거르는 작업 ( 데이타 추출 )

1. 열단위 추출 : df['컬럼명']

2. 행단위 추출 : df.loc[인덱스명칭] / df.iloc[인덱스위치]

    - loc[] : 명칭 기반 인덱싱
    - iloc[] : 위치 기반 인덱싱
    
    현재 인덱스가 글자로 지정되지 않았기에 loc[]/iloc[] 모두 가능
    

        - loc과 iloc 차이
            - df.loc[-1]  # 에러
            - df.iloc[-1] # 마지막번호
